```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('data.csv')
weights = data['weight']

mean_weight = sum(weights) / len(weights)

bins = [30, 40, 50, 60, 70, 80, 90, 100, 110, 120, 130]

plt.hist(weights, bins=bins, edgecolor='black', log=True)

plt.axvline(mean_weight, color='red', label='Weight mean', linewidth=2)
plt.legend()

plt.title('Weights of Respondents')
plt.xlabel('Weights')
plt.ylabel('Total Respondents')

plt.tight_layout()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Seed the random number generator for reproducibility
np.random.seed(1)

# Generate two sets of data points
x = np.random.randn(10000)
y = np.random.randn(10000)

# Create the 2D histogram
plt.hist2d(x, y, bins=30, cmap='Blues')

# Add a color bar to show the intensity scale
plt.colorbar(label='Number of entries')

# Label the axes
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Add a title
plt.title('2D Histogram Example')

# Show the plot
plt.show()
```
